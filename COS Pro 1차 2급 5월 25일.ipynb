위 두 코드셀의 실행 결과에서 알 수 있듯이 먼저 `range(0)`는 일정 범위를 만들지 못했다는 것을 확인했습니다.   

그리고 `for in range(month-1)`은 `month=1`일 때 결국 `for in range(0)`과 같습니다. 결국 `for in` 구조에서 `in`뒤에 이터러블한 데이터가 와서 문법 에러가 나지는 않지만 반복 해서 꺼낼 데이터가 없기 때문에 `for`에 의한 반복 동작이 일어나지 않습니다. 만약 `for`문에 의한 반복 동작이 일어났다면 최소한 한번은 `print('in for')`가 실행되었어야 합니다. 따라서 `total+=month_list[i]` 문장도 실행될 수 없습니다. 

다만 `for`문 밖 `total+=day` 문장부터 실행되어 `total-1`의  값 1이 출력된 것입니다. "1월 2일은 새해 첫날(1월 1일)을 기준으로 1일이 지난 날입니다. 


````python
func_a(1,3)
````

````python
month = 1
day = 3
month_list = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
total = 0;

for i in range(month - 1):
  total += month_list[i]    # 월별 날자
  print('in for')

total += day               # 날자 
print(total-1)
````

func_a(1,3)을 실행하면 1월 3일은 새해 첫날(1월 1일)을 기준으로 2일이 지난 날입니다.

````python
def solution(arr):
    left, right = 0, len(arr)-1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr

#The following is code to output testcase.
arr = [1, 4, 2, 3]
ret = solution(arr)

#Press Run button to receive output.
print("Solution: return value of the function is ", ret, " .")
````

##문제5 분석
* `left, right = 0, len(arr)-1`의 실행은 `left=0`와`right=len(arr)-1`를 실행 한 것과 같습니다. 튜플 언패킹이네요.`len(arr)-1`은 `3`임으로 `right`의 값은 3이 됩니다.

* `while`문 조건 `left < right`에서 조건 `left < right`의 값이 `True`임으로 반복 동작을 시작합니다.

  - `arr[left], arr[right] = arr[right], arr[left]` → `arr[0], arr[3] = arr[3], arr[0]` 맨 뒤 요소(인덱스 3)와 맨 앞 요소(인덱스 0)는 값을 서로 맞교환합니다.
  - `left += 1`와 `right -= 1`연산 후 `left`는 `1`이되고 `right`는 `2`가됨.

* `while`문의 조건 `left < right`이 만족 따라서 다시 반복 연산 수행
  - `arr[left], arr[right] = arr[right], arr[left]` → `arr[1], arr[2] = arr[2], arr[1]` 인덱스 1 요소와 인덱스 2 요소의 값을 서로 맞교환합니다.
   - `left += 1'와 `right -= 1'연산 후 `left`는 `2`가되고 `right`는 `1`이됨.

* `while`문의 조건 `left < right`은 False. 따라서 → `while`문 종료

````python
lst = list(range(1,41))
count = 0

for element in lst:
  quotient = element // 10    # 몫
  remainder = element % 10       # 나머지

  # 경우의 수 4가지 발생
  # 몫 | 나머지 | 출력
  #  F     F      숫자    
  #  F     T      짝
  #  T     F      짝
  #  T     T      짝짝
  
  if quotient not in [3, 6, 9] and remainder not in [3, 6, 9]:     # F F에 해당
    print(element)
  elif quotient in [3, 6, 9] and remainder in [3, 6, 9]:           # T T에 해당
    print('짝짝')
    count += 2
  else:                                                            # 그 외의 경우((T, F) 또는 (F, T))   
    print('짝')
    count += 1
  

print('짝 count :', count)
````
